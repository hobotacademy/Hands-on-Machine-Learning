# 🟣 Function

```python
def calculation():
    a = int(input('Please enter the first number: '))
    b = int(input('Please enter the second number: '))
    c = input('Please enter the operator:')
    
    if c == '+':
        return(a + b)
    elif c == '-':
        print(a - b)
    elif c == '*':
        print(a * b)
    elif c == '/':
        print(a / b)
    else:
        print('Error')
```

```python
y = calculation()
```

```python
y
```

```python
x = lambda a : a + 10

print(x(5)) 
```

## get()

#### get() method takes maximum of two parameters:
#### key - key to be searched in the dictionary
#### value (optional) - Value to be returned if the key is not found. The default value is None.

default − This is the Value to be returned in case key does not exist.

```python
counts = dict()
names = ['aa', 'cc', 'aa', 'z', 'z', 'z', 'z', 'cc', 'aa']
for name in names :
    counts[name] = counts.get(name, 0) + 1
    print(counts)
```

```python
counts
```

```python
counts.items()
```

```python
list(counts.items())
```

```python
l = [(word, count) for word, count in counts.items()]
max(l)
```

```python
max_key = max(counts, key=counts.get)                #########################
print("'{}': {}".format(max_key, counts[max_key]))
```

```python
bigcount = None
bigword = None
for word,count in counts.items():
    if bigcount is None or count > bigcount:
        bigword = word
        bigcount = count

print(bigword, bigcount)
```
