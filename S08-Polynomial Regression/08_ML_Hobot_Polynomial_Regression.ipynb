```python
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([[0],[1],[2],[3],[4]])

y=np.array([[12],[19],[29],[37],[45]])
```

```python
LR = LinearRegression()
```

```python
LR.fit(X, y)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = np.array([[2,3]])
```

```python
poly_conv = PolynomialFeatures(degree=3, include_bias=True)
```

```python
X_poly = poly_conv.fit_transform(x)
```

```python
X_poly
```

```python
x2 = np.array([[2], [3], [4]])
```

```python
poly2 = PolynomialFeatures(degree=3, include_bias=True)
```

```python
xp2 = poly2.fit_transform(x2)
```

```python
xp2
```
