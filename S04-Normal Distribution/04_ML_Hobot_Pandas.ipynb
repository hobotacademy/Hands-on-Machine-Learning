# Q8: What is the average rating of the series?

```python
df2 = pd.read_csv("imdb_top_250_series_episode_ratings.csv")
```

```python
df2.head()
```

```python
df2.drop('Unnamed: 0',axis=1, inplace=True)

#df2 = df2.drop('Unnamed: 0')
```

```python
df2.head()
```

```python
df2.groupby('Title').mean()
```

```python
df_c = df2.groupby(['Title', 'Code']).mean()
```

```python
df_c
```

```python
df_c['Rating']
```

```python
df_c = df_c[['Rating']]
df_c
```

```python
df_c.rename(columns = {'Rating':'Mean Rating'}, inplace = True)
```

```python
df_c.head()
```
