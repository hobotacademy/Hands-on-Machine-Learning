```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def gini(p):
    g = p*(1-(p)) + (1-p)*(1-(1-p))
    
    return g
```

```python
def entropy(p):
    ent = -p* np.log2(p) - (1-p)*np.log2((1-p))
    
    return ent
```

```python
def error(p):
    error = 1- np.max([p, 1-p])
    
    return error
```

```python
x = np.arange(0, 1, 0.001)
```

```python
x.shape
```

```python
# x
```

```python
ent = [entropy(p) if p != 0 else None for p in x]
```

```python
# ent
```

```python
err = [error(i) for i in x]
```

```python
# err
```

```python
fig = plt.figure()

ax = plt.subplot(111)

for i, lab, ls, c, in zip( [ent, gini(x), err], ['Ent', 'Gini', 'Error'],
                         ['-', '-', '-'], ['b', 'g', 'r']):
    line = ax.plot(x, i, label=lab, linestyle=ls, color=c)

```
