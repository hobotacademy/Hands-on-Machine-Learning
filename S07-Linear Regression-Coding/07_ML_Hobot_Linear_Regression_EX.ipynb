# With For

```python
b_init = 785
w_init = np.array([ 0.3, 18.5, -53.3, -26.4])
```

```python
w_init.shape
```

```python
m = X_train.shape[0]

n= X_train.shape[1]
```

```python
y_pred = np.zeros((m,1))

for i in range(m):
    p =0
    for j in range(n):
        p += X_train[i][j] * w_init[j]
    y_pred[i] = p+b_init
```

```python
y_pred
```

```python
y_pred.shape
```
