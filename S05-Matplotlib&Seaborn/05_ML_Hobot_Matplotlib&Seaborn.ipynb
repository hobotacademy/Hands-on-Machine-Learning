# 🟣 Subplots

```python
x = range(10)

y = range(10)

fig, ax = plt.subplots(2)

fig.set_figwidth(10)
fig.set_figheight(10)

for i in ax:
    i.plot(x, y)
```

## m x n subplots

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

plt.subplot(2, 2, 1)
plt.plot(x, y)

plt.subplot(2, 2, 2)
plt.plot(x, y)

plt.subplot(2, 2, 3)
plt.plot(x, y)

plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

for r in ax:
    for c in r:
        c.plot(x, y)

plt.show()
```

## Setting title and labels

```python
x = range(20)
y = range(20)

fig, ax = plt.subplots(nrows=2, ncols=2)

for r in ax:
    for c in r:
        c.plot(x, y)
        c.set_title('T')
        c.set_xlabel('X')
        c.set_ylabel('y')
fig.tight_layout()
        
plt.show()
```

# 🟣 Defining colors

### matplotlib color names
color names

b: blue  
g: green  
r: red  
c: cyan  
m: magenta  
y: yellow  
k: black  
w: white

```python
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black']
```

```python
s = range(1, len(colors)+1)
```

```python
for i, col in zip(s, colors):
    plt.plot([0, 10], [0,i], label=col, lw=3, color=col)
plt.legend(loc='upper left')    
plt.show()
```

# 🟣 Edgecolors for scatter plots

```python
N=50
x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(N)
area = (15*np.random.rand(N))**2
plt.scatter(x, y, s=area, c=colors, alpha=0.5)

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

s = np.random.rand(30, 2)

ax[0][0].scatter(s[:, 0], s[:, 1], color = 'red', label='c=red')

ax[0][1].scatter(s[:, 0], s[:, 1], color = 'blue', label='c=b')

ax[1][0].scatter(s[:, 0], s[:, 1],edgecolor='black', color = 'red', label='c=red')

ax[1][1].scatter(s[:, 0], s[:, 1],edgecolor='black', color = '1', label='c=red')

for r in ax:
    for c in r:
        c.legend(loc='upper left')
plt.show()
```

# 🟣 Line styles

```python
lines = ['-.', '--', '-', '-', ':']

plt.figure(figsize=(8, 5))
s = range(len(lines))

for i in s:
    plt.plot([i-1, i, i+1], [i, i, i],
            label='"%s"' %lines[i],
            linestyle = lines[i],
            lw=4)
plt.legend(loc = 'lower left')    
plt.show()
```

# 🟣 Boxplots

<img src="boxplot_overview.png" alt="Boxplot">

```python
data = [np.random.normal(0, std, 100) for std in range(1, 5)]

fig = plt.figure(figsize=(8, 6))

plt.boxplot(data, notch=True, sym='rs', vert=False)

plt.xticks([y+1 for y in range(len(data))], ['x1', 'x2', 'x3', 'x4'])

plt.show()
```

## Boxplots with custom fill colors

```python
sns.set_style("ticks")

sns.boxplot(data = data)
```

## Violin plots

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

ax[0].violinplot(data, showmeans=False, showmedians=True)

ax[0].set_title('violin plot')

ax[1].boxplot(data)

ax[1].set_title('box plot')

for ax in ax:
    ax.yaxis.grid(True) 
    ax.set_xticks([y+1 for y in range(len(data))], )
    ax.set_xlabel('xlabel')
    ax.set_ylabel('ylabel')

plt.setp(ax, xticks=[y+1 for y in range(len(data))],xticklabels=['x1', 'x2', 'x3', 'x4'])    
    
plt.show()
```
