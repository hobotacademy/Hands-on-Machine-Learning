# Step0: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# Step1: Read Data

```python
df = pd.read_csv("WeatherData.csv")
```

```python
df.head()
```

# Step2: X, y

```python
X = df[['Pressure (millibars)']]
```

```python
y = df[['Humidity']]
```

```python
X = np.array(X)
```

```python
y
```

# Step3: PolynomialFeatures

```python
poly_Feature = PolynomialFeatures(degree=2, include_bias=True)
```

```python
X_poly = poly_Feature.fit_transform(X)
```

```python
df_X_poly = pd.DataFrame(X_poly)
```

```python
df_X_poly.columns = ['1', 'X', 'X^2']

df_X_poly
```

# Step4: Polynomial Regression

## Creat the Model

```python
poly_model = LinearRegression()
```

##  Train the Model

```python
poly_model.fit(X_poly, y)
```

##  Prediction

```python
y_pred = poly_model.predict(X_poly)
```

```python
y_pred.shape
```

```python
X.shape
```

```python
plt.scatter(X, y, c="red")

plt.xlabel("X")

plt.ylabel("y")

plt.plot(X, y_pred)
```

## Evalution

```python
mean_squared_error(y, y_pred)
```

```python
poly_model.intercept_
```

```python
poly_model.coef_
```
