# Naive Bayes

```python
def parameters(X):
    mu=0
    t1=0
    t2=0
    sigma=0
    
    m = X.shape[0]
    mu = (1/m) * X.sum()
    
    t1 = ((X-mu)**2).sum()
    
    t2 = ((X-mu).sum())**2
    
    sigma = (t1 - (t2/m)) / (m-1)
    
    return mu, sigma
```

```python
parameters(X1)
```

```python
def liklihood(x, c):
    mu, sigma = parameters(c)
    t1=0
    t2=0
    
    t1 = 1/(np.sqrt(2*np.pi*(sigma)))
    
    t2 = np.exp(-((x-mu)**2)/(2*sigma))
    
    L = np.array(t1) * np.array(t2)
    
    return L    
```

```python
def prior(df, yc):
    m = df.shape[0]
    prior = yc.shape[0]/m
    
    return prior
```

```python
x_new = [6, 130, 8]
```

```python
def make_d(df, y1, y2, x_new, X1, X2):
    
    p1= prior(df, y1) * np.prod(liklihood(x_new, X1))
    
    p2= prior(df, y2) * np.prod(liklihood(x_new, X2))
    
    if p1>=p2:
        return 1
    
    elif p2>p1:
        return 0
```

```python
make_d(df, y1, y2, x_new, X1, X2)
```
